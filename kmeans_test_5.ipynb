```python
import pandas as pd
video_analiz = pd.read_excel('video_analiz2.xlsx')
video_analiz
```

```python
video_analiz.columns
```

```python
import matplotlib.pyplot as plt

plt.scatter(video_analiz['Video Begeni Sayısı'], video_analiz['Video Yorum Sayısı'])
plt.xlabel('Video Begeni Sayısı')
plt.ylabel('Video Yorum Sayısı')
plt.title('Video Analizi')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(video_analiz['Video Begeni Sayısı'], video_analiz['Video Yorum Sayısı'], color = 'Cyan')
plt.xlabel('Video Begeni Sayısı')
plt.ylabel('Video Yorum Sayısı')
plt.title('Video Analizi')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(video_analiz[['Video Begeni Sayısı', 'Video Yorum Sayısı']])
video_analiz['Küme'] = kmeans.labels_
video_analiz
```

```python
kmeans.cluster_centers_
```

```python
import matplotlib.pyplot as plt

plt.scatter(video_analiz['Video Begeni Sayısı'], video_analiz['Video Yorum Sayısı'], c=video_analiz['Küme'])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red')
plt.xlabel('Video Begeni Sayısı')
plt.ylabel('Video Yorum Sayısı')
plt.title('Video Analizi')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
kmeans.fit(video_analiz[['Video Begeni Sayısı', 'Video Yorum Sayısı']])
video_analiz['Küme'] = kmeans.labels_
video_analiz
```

```python
import matplotlib.pyplot as plt

plt.scatter(video_analiz['Video Begeni Sayısı'], video_analiz['Video Yorum Sayısı'], c=video_analiz['Küme'])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='red')
plt.xlabel("Video Begeni Sayısı (Sayı 15'e bölünmüştür.)")
plt.ylabel('Video Yorum Sayısı')
plt.title('Video Analizi')
plt.show()
```
